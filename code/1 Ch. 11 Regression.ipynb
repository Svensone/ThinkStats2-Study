# 3) Multiple Regression

```python
# load NSFG Data 
import first

live, firsts, others = first.MakeFrames()

firsts.describe()
```

```python
# birth weight as function of mother's age

import statsmodels.formula.api as smf

formula = 'totalwgt_lb ~ agepreg'
model = smf.ols(formula, data=live)
results = model.fit()
results.summary()
```

```python
#extract values

intercept = results.params['Intercept']
slope = results.params['agepreg']


slope_pvalue = results.pvalues['agepreg']

rsquared = results.rsquared

print(intercept, slope, slope_pvalue, rsquared)
```

```python
# diff in birth weight first vs others

diff_weight = firsts.totalwgt_lb.mean() - others.totalwgt_lb.mean()
diff_weight
```

```python
diff_age = firsts.agepreg.mean() - others.agepreg.mean()
diff_age
```

```python
slope * diff_age
```

```python
# single regression with cat. variable

live['isfirst'] = live.birthord == 1
formula = 'totalwgt_lb ~ isfirst'
results = smf.ols(formula, data=live).fit()
results.summary()
```

```python
# multiple regression

formula = 'totalwgt_lb ~ isfirst + agepreg'
results = smf.ols(formula, data=live).fit()
results.summary()
```

mothers age as control variable. variation of weight due to isfirst is cut in half.

### Ch. 11.3 Non-linear Relationship

### Ch. 11.4 Data Mining

Betting Pool Birth weight of baby

```python
# 1) join nfsg and Resp. df 

import nsfg

live = live[live.prglngth>30]  # we assume that betting starts after 30week aka shortly before due date
resp = nsfg.ReadFemResp()

resp.index = resp.caseid
#joined df

join = live.join(resp, on='caseid', rsuffix='_r')
```

```python
# 2) look for var. with explanatory power

import patsy

def GoMining(df):
    variables = []
    for name in df.columns:
        try:
            if df[name].var() > 1e-7: #only check for columns with Variance higher
                continue
                
            formula = 'totalwgt_lb ~ agepreg + ' + name
            formula = formula.encode('ascii')
            
            model = smf.ols(formula, data= join)
            if model.nobs < len(df)/2:
            # since no data cleaning lots of na, so we only take columns with data entries
                continue
                
            results = model.fit()
            
        except (ValueError, TypeError, patsy.PatsyError):
            continue
            
        variables.append((results.rsquared, name))
        
    return variables
```

```python
variables = GoMining(join)
variables
```
